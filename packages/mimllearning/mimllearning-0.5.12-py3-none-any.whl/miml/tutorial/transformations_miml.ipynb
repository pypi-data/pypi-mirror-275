```python
!pip install mimllearning

import pkg_resources
from miml.datasets import load_toy

toy = load_toy()
```

# Original Dataset

```python
toy.show_dataset()
```

# MIMLtoML Transformation

```python
from miml.transformation import ArithmeticTransformation, GeometricTransformation, MinMaxTransformation

print("Arithmetic Transformation")
print("-------------------------")
ArithmeticTransformation().transform_dataset(toy).show_dataset()
print("Geometric Transformation")
print("-------------------------")
GeometricTransformation().transform_dataset(toy).show_dataset()
print("Min Max Transformation")
print("-------------------------")
MinMaxTransformation().transform_dataset(toy).show_dataset()
```

# MIMLtoMI Transformation


```python
from miml.transformation import BinaryRelevanceTransformation, LabelPowersetTransformation

print("Label Powerset Transformation")
print("-----------------------------")

LabelPowersetTransformation().transform_dataset(dataset = toy).show_dataset()

print("Binary Relevance Transformation")
print("-------------------------------")

datasets = BinaryRelevanceTransformation().transform_dataset(dataset = toy)

for i, dataset in enumerate(datasets):
  print("Dataset of label", i+1)
  dataset.show_dataset()

```
